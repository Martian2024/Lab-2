Номер ИСУ - 327164 => Вариант 4<br />
Количество записей, у которых "Название" длиннее 30 символов.

```python
dataset = pd.read_csv('books.csv', encoding='Windows 1251', delimiter=';')
dataset.loc[lambda x: x['Название'].str.len() > 30].size
```

Поиск книги по автору с ограничением выдачи до 200р

```python
author = ''
dataset.loc[(dataset['Автор'].str.contains(author, regex=False, na=False)) & (dataset['Цена поступления'] < 200)]
```

Генератор 20 случайных библиографических ссылок

```python
with open('biblio-links.txt', mode='w') as file:
    for i in range(20):
        line = dataset.iloc[randint(0, dataset.shape[0])]
        file.write(f"{i + 1}. {line['Автор']}. {line['Название']} - {line['Дата поступления']}\n")
```
